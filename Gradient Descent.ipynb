```python
import numpy as np
from numpy.linalg import norm 

def dfdx(x,y):
    return(8*x - 4*y)

def dfdy(x,y):
    return(-4*x + 4*y)

def grad(x,y):
    return np.array([dfdx(x,y), dfdy(x,y)])
```

```python
def gradient(grad, ini):
    
    eps = 0.000001
    max_iter = 10000
    iter = 0
    
    old_min = ini + 10 * eps
    new_min = ini.copy()

    while(norm(new_min - old_min) > eps) and (iter < max_iter):

        alpha = 1/(iter + 1)
        old_min = new_min.copy()
        new_min = old_min - alpha * grad(new_min[0], new_min[1])
        
        print("Iteration", iter + 1, ":", "W is", new_min)
        
        iter += 1
        
    return new_min
```
